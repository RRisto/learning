```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## One iteration

```python
def logistic_iter(x, r):
    "One iteration of the logistic map."
    return r * x - r * x**2
```

```python
logistic_iter(0.1, 2.1)
```

## Multiple iterations

```python
def multiple_iterations_seq(x0, r, n_iters=50):
    values=[logistic_iter(x0, r)]
    for _ in range(n_iters):
        values.append(logistic_iter(values[-1], r))
    return values
```

```python
values=multiple_iterations_seq(0.1, 2.0)
```

```python
plt.plot(values)
```

## Run iters with different r values and see changes

```python
r = np.linspace(0, 4, num=400)
x0 = np.linspace(0, 1, num=400)
vals = np.empty((r.size, x0.size))
for i in range(vals.shape[0]):
    for j in range(vals.shape[1]):
        seq=multiple_iterations_seq(x0[j], r[i])
        vals[i, j]=seq[-1]
```
