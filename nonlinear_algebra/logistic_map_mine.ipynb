```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## One iteration

```python
def logistic_iter(x, r):
    "One iteration of the logistic map."
    return r * x - r * x**2

def logistic_iter(N, r, K=1):
    "population growth sample"
    return  r*N*(1-N/K)
```

```python
logistic_iter(0.1, 2.1)
```

## Multiple iterations

```python
def multiple_iterations_seq(x0, r, n_iters=50):
    values=[logistic_iter(x0, r)]
    for _ in range(n_iters):
        values.append(logistic_iter(values[-1], r))
    return values
```

```python
values=multiple_iterations_seq(0.1, 2.0)
```

```python
plt.plot(values)
```

## Run iters with different r values and see changes

```python
r = np.linspace(0, 4, num=400)
x0 = np.linspace(0, 1, num=400)
vals = np.empty((r.size, x0.size))
for i in range(vals.shape[0]):
    for j in range(vals.shape[1]):
        seq=multiple_iterations_seq(x0[j], r[i])
        vals[i, j]=seq[-1]
```

```python
len([rr * np.ones(vals.shape[0]) for rr in r])#, [row for row in vals]
```

```python
len([row for row in vals])
```

```python
plt.plot([rr * np.ones(vals.shape[0]) for rr in r], [row for row in vals], 'x', ms=0.9)
plt.xlabel('value of $r$ parameter')
```

```python
df_heatmap=pd.DataFrame(vals.T)
df_heatmap.columns=df_heatmap.index/100
df_heatmap.index=(df_heatmap.index/400)[::-1]
sns.heatmap(df_heatmap, robust=True, cmap='Blues')
```

## Logistic iter solution 

```python
f = lambda r,x: r * x - r * x**2
x= np.linspace(0,100,100) #population 
r=np.linspace(0,1,100) #carrying capacity, how much population we can carry with exsiting resources
r_hat=f(r, x)
```

```python
plt.plot(r, r_hat)
plt.hlines(0, 0, r.max(), color='red')
```
