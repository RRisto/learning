```python
import numpy as np
import matplotlib.pyplot as plt
```

## Population growth example

#### from Nonlinear Dynamics And Chaos: With Applications To Physics, Biology, Chemistry, And Engineering page 31 (Population Growth)

```python
f = lambda N,K: N*(1-N/K) #omitted r
N= np.linspace(0,100,100) #population 
K=90 #carrying capacity, how much population we can carry with exsiting resources
N_hat=f(N, K)
```

```python
plt.plot(N, N_hat)
plt.hlines(0, 0, N.max(), color='red')
```
