```python
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
import pickle
import os
from pathlib import Path
```

```python
# df = pd.read_hdf("D:\Mtech\Sem4\ASR\Project\Feature\\timit_mfcc_delta_delta.hdf")
df = pd.read_hdf("./features/mfcc/timit.hdf")
features = np.array(df["features"].tolist())
label = np.array(df["labels"].tolist())
df1 = pd.DataFrame(features)
df2 = pd.DataFrame(label)
df = pd.merge(df1, df2, right_index=True, left_index=True)
df.rename(columns={'0_y': 'Label'}, inplace=True)
print(df.head())
```

```python
labels=df.Label.unique()
print(labels)
print(len(labels))
labels= pd.Series(labels)
for folder in ['models/2M_GMM','models/4M_GMM', 'models/8M_GMM','models/16M_GMM',
               'models/32M_GMM','models/64M_GMM','models/128M_GMM']:
    Path(folder).mkdir(exist_ok=True)
```

```python
def train_gmm(x):
    print(type(x),x)
    global df
    for i in range(1,8):
        array = df.loc[df['Label']==x].iloc[:,:-1]
        gmm = GaussianMixture(n_components=2**i,covariance_type='diag').fit(array)
        file = ''.join(['models/',str(2**i),'M_GMM\\',x,'_', str(2**i) ,'.pkl'])
        print(file)
        with open(file, 'wb') as f:
            pickle.dump(gmm, f)

labels.apply(train_gmm)
```
