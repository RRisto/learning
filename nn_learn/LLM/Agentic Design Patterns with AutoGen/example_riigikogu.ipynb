## helpers

```python
def read_df(path='stenos_cleaned_faction2022_subset.xlsx'):
    import pandas as pd
    df=pd.read_excel(path)
    df['datetime']=pd.to_datetime(df.date)
    return df
```

```python
def count_of_events(text: str) -> str:
    """Given a text string finds number of events and returns the answer."""
    df=read_df()
    n=df[df.text.str.contains(text)].shape[0]
    return f"Number of events containing '{text}' is: {n}"


def changes_in_events(text_period1_start: str, text_period1_end: str, text_period2_start:str, text_period2_end:str) -> str:
    """Compares changes in number of events in two time periods. Makes two queries: text_period1_start and  
    text_period1_end for first period and text_period2_start and text_period2 for second period.
    Periods must be in format YYYY-MM-DD"""
    
    df=read_df()
    nodes_period1=df[df.datetime>=pd.to_datetime(text_period1_start)&(df.datetime<=pd.to_datetime(text_period1_end))]
    nodes_period2=df[df.datetime>=pd.to_datetime(text_period2_start)&(df.datetime<=pd.to_datetime(text_period2_end))]
    change_factor=len(nodes_period2)/len(nodes_period1)
    return f"Most significant change is that number of events changes by a factor of {change_factor}"


def show_sample_events(text: str, n:int=100) -> str:
    """Given a text string finds events and returns n of them"""
    
    df=read_df()
    df_=df[df.text.str.contains(text)]
    
    result=f"Events which contain '{text}' are: {[text for text in df_.text.tolist()[:n]]}".encode('ascii', 'replace').decode('ascii')
    print(result)
#     with open('result.txt','w', encoding='utf-8') as f:
#         f.write(result, f)
    return f"Events which contain '{text}' are: {[text for text in df_.text.tolist()[:n]]}"

def plot_events_timeline(text: str, n:int=100, time_granularity:str='day') -> str:
    """Given a text string finds events and their count distribution on timeline and plots it.
    'n' is the number of nodes to consider, by default 100
    'time_granularity' is granularity of time for plotting, must be one of: 'day', 'week' or 'month'
    Returns message plot is generated and matplotlib plot object"""
    import matplotlib.pyplot as plt
    
    df=read_df()
    if time_granularity not in ['day', 'week', 'month']:
        time_granularity='day'
        
    df_=df[df.text.str.contains(text)]
    plot=df_.groupby(['year', time_granularity])['index_snd'].count().unstack().T.plot()
    plt.savefig('plot.png')
    plt.show()
    
    return "plot is generated and visible here:temp/plot.png", plot

def plot_events_metadata_distribution(text: str, metadata_key:str, n:int=100) -> str:
    """Given a text string finds events and counts their metadata_key distribution and plots it.
    'n' is the number of nodes to consider, by default 100
    'metadata_key' is metadata column that is used for dividing data into groups"""
    
    df=read_df()
        
    df_=df[df.text.str.contains(text)]
    plot=df_[metadata_key].value_counts().plot(kind='barh')
    plt.savefig('plot.png')
    plt.show()
    
    return "plot is generated and visible here:temp/plot.png"
```
