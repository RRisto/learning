```python
import requests
import pandas as pd
import json
```

## Data

```python
reps=requests.get('https://rata.digitraffic.fi/api/v1/live-trains/station/HKI?arrived_trains=5&departed_trains=10&include_nonstopping=False')
```

```python
jsn=json.loads(reps.content.decode('utf-8'))
```

```python
df=pd.DataFrame(jsn)
df.head()
```

```python
pd.DataFrame(df.timeTableRows)
```

```python
df_timetable=pd.DataFrame()

for i in range(df.shape[0]):
    df_temp=pd.DataFrame(df.timeTableRows[i])
    df_temp['trainNumber']=df.trainNumber[i]
    df_timetable=df_timetable.append(df_temp)
df_timetable=df_timetable.reset_index(drop=True)
df_timetable.shape
```
